```
# Define function to categorize PRH4 items
def categorize_prh4(prh4):
    if pd.isna(prh4):
        return 'Unknown'
    if 'Implants' in prh4:
        return 'Implants'
    elif 'Abutments' in prh4:
        return 'Abutments'
    elif 'Surgical' in prh4 or 'Healing' in prh4:
        return 'Surgical and Healing Products'
    elif 'Prosthetics' in prh4 or 'Restorative' in prh4:
        return 'Prosthetics and Restorative Solutions'
    elif 'Instruments' in prh4 or 'Accessories' in prh4:
        return 'Instruments and Accessories'
    elif 'Software' in prh4 or 'Services' in prh4:
        return 'Software and Services'
    else:
        return 'Miscellaneous'

# Apply the categorization function to the PRH4 column
df['PRH4_Category'] = df['PRH4'].apply(categorize_prh4)

# Display the DataFrame with the category column
print(df)
```

```
df['order_date'] = pd.to_datetime(df['order_date'])

# Generate a new column 'order_year' containing the year
df['order_year'] = df['order_date'].dt.year
```

```
df.head()
```

```
# @title
# Define the categorization mapping
categorization = {
    'Academic Institutions': ['University', 'Institution'],
    'Clinical Practitioners': ['General dentist', 'Periodontist', 'Surgeon (gen./oral)',
                               'Prosthodontist', 'Orthodontist', 'Maxillofacial surg'],
    'Laboratories': ['Laboratory', 'In-house Laboratory'],
    'Healthcare Facilities': ['Hospital/Clinic'],
    'Specialized Roles': ['Denturist', 'Central services', 'Staff'],
    'Government and Military': ['Government/Military'],
    'Business Entities': ['Distributor/Reseller'],
    'Employment Categories': ['Employee'],
    'Other': ['NOT MAPPED', 'Others', np.nan]
}

# Function to categorize the customer group
def categorize(group):
    for category, groups in categorization.items():
        if group in groups:
            return category
    return 'Other'

# Apply the categorization function to the DataFrame
df['customer_category'] = df['customer_group'].apply(categorize)

# Display the categorized DataFrame
print(df)
```

```
df['customer_category'].unique()
```

```
# Group by 'order_year' and 'sales_channel', then count unique customers and sum net value
sales_customers = df.groupby(['order_year', 'sales_channel']).agg({'order_number': 'size','customer_number': 'nunique', 'order_material_net_value': 'sum'}).reset_index()
sales_customers.rename(columns={'customer_number': 'unique_customers_count', 'order_material_net_value': 'total_net_value'}, inplace=True)

# Calculate average order value
sales_customers['average_order_value'] = sales_customers['total_net_value'] / sales_customers['unique_customers_count']

# Filter for E-shop and Non-Eshop
e_shop_counts = sales_customers[sales_customers['sales_channel'] == 'E-shop']
non_e_shop_counts = sales_customers[sales_customers['sales_channel'] == 'Non-Eshop']
```

```
# Group by 'order_year' and 'sales_channel', then count unique customers and sum net value
sales_customers_PRH = df[df['order_year']!=2024].groupby(['order_year', 'sales_channel','PRH1']).agg({'order_number': 'size','customer_number': 'nunique', 'order_material_net_value': 'sum'}).reset_index()
sales_customers_PRH.rename(columns={'order_number':'order_count','customer_number': 'unique_customers_count', 'order_material_net_value': 'total_net_value'}, inplace=True)

# Calculate average order value
sales_customers_PRH['average_customer_value'] = sales_customers_PRH['total_net_value'] / sales_customers_PRH['unique_customers_count']
sales_customers_PRH['average_order_value'] = sales_customers_PRH['total_net_value'] / sales_customers_PRH['order_count']

# Filter for E-shop and Non-Eshop
e_shop_counts_PRH = sales_customers_PRH[sales_customers_PRH['sales_channel'] == 'E-shop']
non_e_shop_counts_PRH = sales_customers_PRH[sales_customers_PRH['sales_channel'] == 'Non-Eshop']
```

```
# Group by 'order_year' and 'sales_channel', then count unique customers and sum net value
sales_customers_PRH4 = df[df['order_year']!=2024].groupby(['order_year', 'sales_channel','PRH4_Category']).agg({'order_number': 'size','customer_number': 'nunique', 'order_material_net_value': 'sum'}).reset_index()
sales_customers_PRH4.rename(columns={'order_number':'order_count','customer_number': 'unique_customers_count', 'order_material_net_value': 'total_net_value'}, inplace=True)

# Calculate average order value
sales_customers_PRH4['average_customer_value'] = sales_customers_PRH4['total_net_value'] / sales_customers_PRH4['unique_customers_count']
sales_customers_PRH4['average_order_value'] = sales_customers_PRH4['total_net_value'] / sales_customers_PRH4['order_count']

# Filter for E-shop and Non-Eshop
e_shop_counts_PRH4 = sales_customers_PRH4[sales_customers_PRH4['sales_channel'] == 'E-shop']
non_e_shop_counts_PRH4 = sales_customers_PRH4[sales_customers_PRH4['sales_channel'] == 'Non-Eshop']
```
